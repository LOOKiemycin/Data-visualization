```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
df = pd.read_csv('Data.csv')
df
```

# secondary y-axis using the same x-axis

```python
fig, ax = plt.subplots(figsize=(15,8))
ax.plot(df.Time, df.Signal, color='black')
ax.set_xlabel("Retention time (min)", color="black",fontsize=16)
ax.set_ylabel("Absorbance (mAU)", color="black",fontsize=16)
ax.set_xlim(0, 12.5)
ax.tick_params(labelsize = 14)
ax.set_ylim(-10, 1500)

ax2 = ax.twinx()

Time = df.Time
Inhibition = df.Inhibition
  
ax2.bar(Time, Inhibition, color="lightgray", alpha=0.4, width=0.125, ec='red')
ax2.set_ylabel("Inhibition (%)", color="black",fontsize=16)
ax2.tick_params(labelsize = 14)
ax2.set_ylim(0, 100)
plt.xlim(0,12.5)

for i in range(len(Inhibition)):
    
# Use X value as label and format number
    label = '{:,.1f}'.format(Inhibition[i])
    
    plt.annotate(label, 
                 xy=(Time[i],Inhibition[i]),
                 ha='center', rotation=90, va='bottom',
                 size=15, xytext=(0, 5),
                 textcoords='offset points')

#plt.savefig('Filename.png', bbox_inches='tight', dpi=300)
```
